```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Salary_dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
data= df.drop(['Unnamed: 0'],axis=1)
```

```python
corr= data.corr()
```

```python
sns.heatmap(corr,vmin=-1, vmax=1,cmap='coolwarm',annot=True)
plt.title('Correlation of salary and experience')
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
data['YearsExperience'] = round(data['YearsExperience'],0)
```

```python
pivoted=data.pivot_table(columns='YearsExperience',values='Salary',aggfunc='sum')
```

```python
sns.barplot(pivoted)
plt.title('Experience and Salary')
plt.xlabel('Number of Experience')
plt.ylabel('Salary')
plt.xticks(rotation=50)
plt.show()
```

```python
sns.histplot(data['Salary'],kde=True)
plt.title('Salary Distribution')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df[['YearsExperience']]
y= df['Salary']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=30)
```

```python
 model= LinearRegression().fit(X_train,y_train)
```

```python
model
```

```python
prediction= model.predict(X_test)
```

```python
prediction
```

```python
plt.scatter(X_train,y_train,alpha=0.5,marker='o',c='blue')
plt.plot(X_train,model.predict(X_train),c='red',lw=0.2)
plt.title('Salary and experience')
plt.ylabel('Salary')
plt.xlabel('Number of Experience')
plt.show()
```
